```python
BRAIN_FILE="/content/drive/My Drive/Colab Notebooks/aiml_pretrained_model.dump"

```

```python
k = aiml.Kernel()

```

```python
if os.path.exists(BRAIN_FILE):
  print("Loading from brain file: " + BRAIN_FILE)
  k.loadBrain(BRAIN_FILE)
else:
  print("Parsing aiml files")
  k.bootstrap(learnFiles="/colearningFve/Colab Notebooks/learningFileList.aiml", commands="load aiml")
  print("Saving brain file: " + BRAIN_FILE)
  k.saveBrain(BRAIN_FILE)

```

```python

while True:
    query = input("User > ")
    query = [spell(w) for w in (query.split())]
    question = " ".join(query)
    response = k.respond(question)
    if response:
        print("bot > ", response)
    else:
        print("bot > :) ", )
```
